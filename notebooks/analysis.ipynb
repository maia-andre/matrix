```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../datasets/seed7.csv')
df.tail()
```

```python
# demografia: populacao, energia media e comida total ao longo dos ticks
fig, eixos = plt.subplots(3, 1, figsize=(10, 7), sharex=True)
eixos[0].plot(df['tick'], df['pop'])
eixos[0].set_ylabel('pop')
eixos[1].plot(df['tick'], df['energia_media'])
eixos[1].set_ylabel('energia media')
eixos[2].plot(df['tick'], df['comida_total'])
eixos[2].set_ylabel('comida total')
eixos[2].set_xlabel('tick')
fig.suptitle(f"seed {df['seed'].iloc[0]}")
plt.tight_layout()
plt.show()
```

```python
# bateria de desbotamento: os 5 mostradores, todos em [0,1]
mostradores = ['modelo', 'agencia', 'modelo_do_outro', 'phi', 'relato']
fig, eixo = plt.subplots(figsize=(10, 4))
for m in mostradores:
    eixo.plot(df['tick'], df[m], label=m)
eixo.set_xlabel('tick')
eixo.set_ylim(0, 1)
eixo.legend()
eixo.set_title('bateria de desbotamento')
plt.tight_layout()
plt.show()
```

```python
# evolucao dos tracos (nivel 6): media da populacao com banda de +-1 desvio
tracos = [('hor', 'HORIZONTE'), ('desc', 'DESCONTO'), ('urg', 'URGENCIA'), ('esp', 'PESO_ESPACO')]
fig, eixos = plt.subplots(2, 2, figsize=(10, 6), sharex=True)
for eixo, (t, nome) in zip(eixos.flat, tracos):
    m, sd = df[f'{t}_m'], df[f'{t}_sd']
    eixo.plot(df['tick'], m)
    eixo.fill_between(df['tick'], m - sd, m + sd, alpha=0.25)
    eixo.set_title(nome)
plt.tight_layout()
plt.show()
```
